```python
import pandas as pd
```

```python
# From dictionary
data = {'Name': ['Alice', 'Bob'], 'Age': [25, 30]}
df = pd.DataFrame(data)
 
# From list of lists
lst = [['Alice', 25], ['Bob', 30]]
df = pd.DataFrame(lst, columns=['Name', 'Age'])
```

```python
df['Gender'] = ['F', 'M']  # Add new column
df.drop('Age', axis=1, inplace=True, errors='ignore')  # Remove column if it exists
```

```python
# Merge two DataFrames
df1 = pd.DataFrame({'key': df['Name'], 'value1': df['Age']})
df2 = pd.DataFrame({'key': df['Name'], 'value2': df['Gender']})

pd.merge(df1, df2, on='key')

# Join by index
df1.set_index('key').join(df2.set_index('key'))
 
# Join by index
df1.set_index('key').join(df2.set_index('key'))
```
